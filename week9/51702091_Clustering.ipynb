```python
def distance(a, b):
    return math.sqrt((a[0] - b[0])**2 + (a[1] - b[1])**2)

list_map = list(map_dict.items()) 

def choose_center_first(set_point): 
    a = random.randint(0,len(set_point)-1)
    b = a
    while a == b:
        b = random.randint(0,len(set_point)-1)
    return a, b


# def choose_center(set_point): 
cluster={}

def update_center(set_point):  
    x = list(map(lambda x:x[0],set_point))
    y = list(map(lambda x:x[1],set_point)) 
    return [np.mean(x),np.mean(y)]

def update_cluster(a, b):
    for i in range(2): 
        cluster[i] = [] 
    for item in list_map:
        dis_1 = distance(a,item[1]) 
        dis_2 = distance(b,item[1])
        if dis_1 < dis_2:
            cluster[0].append(item[1])
        else: 
            cluster[1].append(item[1]) 
    return cluster

def partition(k=2):
    index_a, index_b = 23, 13 #choose_center_first(map_dict)   
    new_a = list(map_dict.items())[0][1]
    new_b = list(map_dict.items())[1][1]
    a= new_a
    b= new_b   
    print(len(map_dict.items()))
    ll = list(map(lambda x:x[1],list(map_dict.items())))
    plt.scatter(list(map(lambda x:x[0],ll)),list(map(lambda x:x[1],ll)),c='b')
    plt.scatter(new_a[0],new_a[1],c='yellow')
    plt.scatter(new_b[0],new_b[1],c='purple')
    plt.show()
    while(True):  
        cluster = update_cluster(new_a, new_b)
         
        center_1 = update_center(cluster[0])
        center_2 = update_center(cluster[1])   
        if abs(np.sum(np.array(center_1) -np.array(new_a))) < 10e-9 and abs(np.sum(np.array(center_2) -np.array(new_b))) < 10e-9:
            break 
        new_a = center_1
        new_b = center_2
        x_1 = list(map(lambda x:x[0],cluster[0]))
        y_1 = list(map(lambda x:x[1],cluster[0]))
        x_2 = list(map(lambda x:x[0],cluster[1]))
        y_2 = list(map(lambda x:x[1],cluster[1]))
        plt.scatter(x_1,y_1,c='r')
        plt.scatter(x_2,y_2,c='b') 
        print(new_a,new_b)
        plt.scatter(new_a[0],new_a[1],c='yellow')
        plt.scatter(new_b[0],new_b[1],c='purple')
        plt.show()

partition()
```
